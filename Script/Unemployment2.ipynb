```python
# Concatanate the all the states into one dataframe

df_data_combined_states = pd. DataFrame() 
df_data_combined_states = pd.concat([df_data_alaska, df_data_california, df_data_hawaii, df_data_newyork, df_data_tenn, df_data_texas, df_data_washington])
```

```python
df_data_combined_states = df_data_combined_states.reset_index(drop = True)
```

```python
# series table has all the discriptions of all the data that is stored in a state file 
df_ref_series = pd.read_csv('C:/Users/Kiran/Desktop/NSS/Capstone/data/laus/ref_series',sep='\t',header=(0))
df_ref_series.columns = df_ref_series.columns.str.lower().str.strip()
df_ref_series.head(5)
```

```python
# Measure text shows type of employment measures captured
df_ref_measure = pd.read_csv('C:/Users/Kiran/Desktop/NSS/Capstone/data/laus/ref_measure',sep='\t',header=(0))
df_ref_measure.columns = df_ref_measure.columns.str.lower().str.strip()
df_ref_measure
```

```python
# has the description for period as month names

df_ref_period = pd.read_csv('C:/Users/Kiran/Desktop/NSS/Capstone/data/laus/ref_period',sep='\t',header=(0))
df_ref_period.columns = df_ref_period.columns.str.lower().str.strip()
df_ref_period
```

```python
# description for seasonal codes

df_ref_seasonal = pd.read_csv('C:/Users/Kiran/Desktop/NSS/Capstone/data/laus/ref_seasonal',sep='\t',header=(0))
df_ref_seasonal.columns = df_ref_seasonal.columns.str.lower().str.strip()

df_ref_seasonal
```

```python
# Name discription for State codes

df_ref_states = pd.read_csv('C:/Users/Kiran/Desktop/NSS/Capstone/data/laus/ref_states',sep='\t',header=(0))
df_ref_states.columns = df_ref_states.columns.str.lower().str.strip()
df_ref_states
```

```python
# -- This is at the city level - come to this later

df_data_city = pd.read_csv('C:/Users/Kiran/Desktop/NSS/Capstone/data/laus/data_city',sep='\t',header=(0))
df_data_city.columns = df_data_city.columns.str.lower().str.strip()
df_data_city
```

```python
# description of area type codes - state wide; Metro area etc
df_ref_area_type = pd.read_csv('C:/Users/Kiran/Desktop/NSS/Capstone/data/laus/ref_area_type',sep='\t',header=(0))
df_ref_area_type.columns = df_ref_area_type.columns.str.lower().str.strip()
df_ref_area_type
```

```python
# merging all states data with various description tables
df_data_combined_states = pd.merge(df_data_combined_states, df_ref_series, on = 'series_id', how = 'inner', 
                                   suffixes=('__df_data_combined_states', '__df_ref_series'))
```

```python
# Bringing in State Names
df_ref_states
df_data_combined_states = pd.merge(df_data_combined_states, df_ref_states, on = 'srd_code', how = 'inner', suffixes=('__df_data_combined_states', '__df_ref_states'))
df_data_combined_states.head(2)
```

```python
#Bringing in measure text description
df_ref_measure
df_data_combined_states = pd.merge(df_data_combined_states, df_ref_measure, on = 'measure_code', how = 'inner', suffixes=('__df_data_combined_states', '__df_ref_measure'))
df_data_combined_states.head()
```

```python
# Bringing in month Names
df_ref_period
df_data_combined_states = pd.merge(df_data_combined_states, df_ref_period, on = 'period', how = 'inner', suffixes=('__df_data_combined_states', '__df_ref_period'))
df_data_combined_states.head()
```

```python
# Bringing in area desciption

df_ref_area_type
df_data_combined_states = pd.merge(df_data_combined_states, df_ref_area_type, on = 'area_type_code', how = 'inner', suffixes=('__df_data_combined_states', '__df_ref_area_type'))
```

```python
# Bringing in season description

df_ref_seasonal
df_data_combined_states = pd.merge(df_data_combined_states, df_ref_seasonal, left_on='seasonal', right_on='seasonal_code', how = 'inner', suffixes=('__df_data_combined_states', '__df_ref_seasonal'))
df_data_combined_states
```
