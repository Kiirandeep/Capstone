```python
# Bringing in month Names
df_ref_period
df_data_combined_states = pd.merge(df_data_combined_states, df_ref_period, on = 'period', how = 'inner', suffixes=('__df_data_combined_states', '__df_ref_period'))
df_data_combined_states.head()
```
