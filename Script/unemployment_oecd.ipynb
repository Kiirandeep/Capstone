```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline 
```

```python
oecd_all = pd.read_csv('../data/MIG_NUP_RATES_GENDER_12042024205534698.csv')
```

```python
#dropping repeated cols
oecd_all1 = oecd_all[['Country','Place of birth', 'Gender', 'Rate', 'Year', 'Unit','Value']]
```

```python
oecd_us = oecd_all1.loc[oecd_all1.Country == 'United States']
oecd_us = oecd_us.loc[(oecd_us.Rate== 'Unemployment rate') & (oecd_us.Gender == 'Total')].reset_index(drop = True)
us_df = oecd_us.groupby('Year')['Value'].sum().to_frame().reset_index()
us_df
```

```python
oecd_us.info()
```

```python
plt.figure(figsize = (15,5))
plt.plot(us_df['Year'],us_df['Value']);
```
