```python
import pandas as pd
```

```python
states = pd.read_csv('states.csv')
```

```python
states
```

```python
for ind, value in states.iterrows():
    if value.year != 2010:
        pct_drop = 100 - (value.value/states.loc[ind-1, 'value'] *100)
        states.loc[ind, 'pct_drop'] = pct_drop
    else:
        states.loc[ind, 'pct_drop'] = 0
```
